```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Churn.csv")
```

**To view unique items**

```python
df['state'].unique()
```

**To get the number of unique items**

```python
df['state'].nunique()
```

**To get the number of occurences of each Items**

```python
df['state'].value_counts().head()
```

```python
df['state'].value_counts(normalize=True).head()
```

**get the first 'n' rows order by a given column**

```python
df.nlargest(5, "account length")
```

**Drop a column from a DataFrame**

Use inplace=True to effect in the dataframe itself

```python
df.drop(columns="churn")
```

```python
df.head()
```

**Drop record by Index**

```python
df.drop(index=[0,1,2])
```

Transpose a DataFrame

```python
df.T
```

```python
df.T.info()
```

***Display Options***

```python
# Reset
pd.reset_option("display.max_rows")
pd.reset_option("display.max_columns")
pd.reset_option("display.precision")
pd.reset_option("display.float_format")
```

**By defaults, pandas truncates rows and columns**

```python
arr = np.random.randint(1, 100, (80, 25))
```

```python
df = pd.DataFrame(arr)
```

```python
df
```

**Print the current value of settings**

```python
# Options
print(pd.options.display.max_rows)
print(pd.options.display.max_columns)
```

**CAVEAT**: The max_rows settings only displays more rows, If the setting is greater than the number of rows in the DataFrame

```python
pd.set_option("display.max_rows", 100)
pd.set_option("display.max_columns", 30)
df
```
